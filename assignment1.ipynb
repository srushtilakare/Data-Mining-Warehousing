```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset 

```python
#import dataset
df=pd.read_csv("heart.csv")
```

# Description of each column:(14 features)

1] age: Age of the patient in years.

2] Sex: Gender of the patient (1 = male; 0 = female).

3] cp: Type of chest pain experienced, categorized as:

0: Typical angina
1: Atypical angina
2: Non-anginal pain
3: Asymptomatic

4] trestbps: Resting blood pressure in mm Hg upon hospital admission.

5] chol: Serum cholesterol level in mg/dL.

6] fbs: Fasting blood sugar > 120 mg/dL (1 = true; 0 = false).

7] restecg: Resting electrocardiographic results, with values:

0: Normal
1: Having ST-T wave abnormality
2: Showing probable or definite left ventricular hypertrophy

8] thalach: Maximum heart rate achieved during exercise.

9] exang: Exercise-induced angina (1 = yes; 0 = no).

10] oldpeak: ST depression induced by exercise relative to rest.

11] slope: Slope of the peak exercise ST segment, categorized as:

0: Upsloping
1: Flat
2: Downsloping

12] Ca: Number of major vessels (0-3) colored by fluoroscopy.

13] thalassemia: Thalassemia status, with values:

0: Normal
1: Fixed defect
2: Reversible defect

14] target: Presence of heart disease (1 = yes; 0 = no).

# Central Tendency

```python
#To find mean we used mean()
mean_age=df['age'].mean()
print(f'Mean Age of Patients :{mean_age}')
```

```python
#To find median we used median()
median_age=df['age'].median()
print(f'Median Age of Patients :{median_age}')
```

```python
#To find mode we used mode()
mode_age=df['age'].mode()
print(f'Mode Age of Patients :{mode_age}')
```

```python
# To find variance we used var()
variance_age=df['age'].var()
print(f'Variance Age of Patients :{variance_age}')
```

```python
#To find standard deviation we used std()
standard_age=df['age'].std()
print(f'Standard deviation Age of Patients :{standard_age}')
```

# Box and Quantitle Plot

```python
#plot boxplot to visulize outliers
plt.figure(figsize=(7,5))
for column in df.select_dtypes(include=np.number).columns:
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot for {column}')
    plt.show()
```

```python
q1=df['age'].quantile(0.25)
q3=df['age'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q1+1.5*iqr
df['age']=np.clip(df['age'],lower_bound,upper_bound)
```

```python
# Select a numerical column (e.g., 'age')
numerical_column = 'age'
quartiles = df[numerical_column].quantile([0.25, 0.5, 0.75])
print(quartiles)
```

```python
#Q-Q plot using scipy's probplot
import scipy.stats as stats
plt.figure(figsize=(12,5))
for column in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(10,5))
    stats.probplot(df[column].dropna(), dist="norm", plot=plt)
    plt.title(f'Q-Q plot for {column}')
    plt.show()
```

```python
# 3D Box Plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['age'], df['chol'], df['trestbps'], c='blue', alpha=0.6)
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
ax.set_zlabel("Resting Blood Pressure")
plt.title("3D Scatter Plot")
plt.show()
```

# Scatter Plot

```python
# Scatter plot between age and cholesterol
plt.figure(figsize=(7, 5))
plt.scatter(df['age'], df['chol'], alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.title("Scatter plot of Age vs Cholesterol")
plt.show()
```

# Histogram

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['chol'], bins=30, kde=True, color='g')
plt.xlabel("Cholesterol Level")
plt.ylabel("Frequency")
plt.title("Histogram: Cholesterol Distribution")
plt.show()
```
